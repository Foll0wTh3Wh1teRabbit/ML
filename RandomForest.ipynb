Случайный лес — это контролируемый алгоритм машинного обучения, используемый для классификации, регрессии и других задач с использованием деревьев решений. Случайные леса особенно хорошо подходят для обработки больших и сложных наборов данных, работы с многомерными пространствами объектов и предоставления информации о важности объектов. Способность этого алгоритма поддерживать высокую точность прогнозирования при минимизации переобучения делает его популярным выбором среди других в различных областях.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, precision_score, recall_score, mean_absolute_error, mean_squared_error
from sklearn import tree
```

```python
iris = load_iris()

df = pd.DataFrame(iris.data, columns = iris.feature_names)
df["target"] = iris.target

df
```

```python
X, y = load_iris(return_X_y = True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)

fn=iris.feature_names
cn=iris.target_names
```

```python
_, ax = plt.subplots()
scatter = ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train)

ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])

_ = ax.legend(scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes")
```

```python
rfc = RandomForestClassifier().fit(X_train[:, :2], y_train)
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 3,figsize = (10,2), dpi=900)

# визуализация 3 деревьев решений из ансамбля, на деле их больше
for index in range(0, 3):
    tree.plot_tree(rfc.estimators_[index],
                   feature_names = fn,
                   class_names=cn,
                   filled = True,
                   ax = axes[index]);

    axes[index].set_title('Estimator: ' + str(index), fontsize = 11)
```

```python
disp = DecisionBoundaryDisplay.from_estimator(
    rfc,
    X_train[:, :2],
    response_method="predict",
    xlabel=iris.feature_names[0], ylabel=iris.feature_names[1],
    alpha=0.5,
    cmap=plt.cm.coolwarm
)

# Plotting the data points
disp.ax_.scatter(X_train[:, 0], X_train[:, 1],
                 c=y_train, edgecolor="k",
                 cmap=plt.cm.coolwarm)

plt.title(f"Decision surface for tree trained on {iris.feature_names[0]} and {iris.feature_names[1]}")
plt.show()
```

```python
y_pred = rfc.predict(X_test[:, :2])

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, average='macro'))
print("Recall:", recall_score(y_test, y_pred, average='macro'))
```

```python
clf = RandomForestClassifier().fit(X_train, y_train)

fig, axes = plt.subplots(nrows = 1,ncols = 3,figsize = (10,2), dpi=900)

# визуализация 3 деревьев решений из ансамбля, на деле их больше
for index in range(0, 3):
    tree.plot_tree(rfc.estimators_[index],
                   feature_names = fn,
                   class_names=cn,
                   filled = True,
                   ax = axes[index]);

    axes[index].set_title('Estimator: ' + str(index), fontsize = 11)

plt.show()
```

```python
y_pred = clf.predict(X_test)
```

```python
cf_mat = confusion_matrix(y_test, y_pred)

cm_fig, cm_ax = plt.subplots(figsize=(3, 3))
cf_mat_disp = ConfusionMatrixDisplay(cf_mat)
cf_mat_disp.plot(ax=cm_ax, cmap=plt.cm.Reds)

plt.show()
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, average='macro'))
print("Recall:", recall_score(y_test, y_pred, average='macro'))
```
