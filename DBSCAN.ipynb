Density-based spatial clustering of applications with noise, DBSCAN — это алгоритм кластеризации данных, который предложили Маритин Эстер, Ганс-Петер Кригель, Ёрг Сандер и Сяовэй Су в 1996. Это алгоритм кластеризации, основанной на плотности — если дан набор точек в некотором пространстве, алгоритм группирует вместе точки, которые тесно расположены (точки со многими близкими соседями), помечая как выбросы точки, которые находятся одиноко в областях с малой плотностью (ближайшие соседи которых лежат далеко). DBSCAN является одним из наиболее часто используемых алгоритмов кластеризации, и наиболее часто упоминается в научной литературе.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
```

```python
centers = [[1, 1], [-1, -1], [1, -1]]
X, y_true = make_blobs(n_samples=500, centers=centers, cluster_std=0.4, random_state=40000)
```

```python
plt.figure(figsize=(12,9))

plt.scatter(X[:, 0], X[:, 1], s=50)

plt.show()
```

Попробуем в лоб использовать алгоритм

```python
predicted = DBSCAN().fit_predict(X)

predicted
```

```python
print(f"Detected clusters: {len(np.unique(predicted))}, actual: 3+1")
```

```python
plt.figure(figsize=(12,9))

plt.scatter(X[:, 0], X[:, 1], c=predicted, s=50)

plt.show()
```

Получилось плохо... Попробуем поиграться с параметрами

```python
new_predicted = DBSCAN(eps=0.4, min_samples=30).fit_predict(X)

new_predicted
```

```python
print(f"Detected clusters: {len(np.unique(new_predicted))}, actual: 3+1")
```

```python
plt.figure(figsize=(12,9))

plt.scatter(X[:, 0], X[:, 1], c=new_predicted, s=50)

plt.show()
```

```python
plt.scatter(X[new_predicted==-1, 0], X[new_predicted==-1, 1], c='green', s=70, marker='x')
plt.scatter(X[new_predicted!=-1, 0], X[new_predicted!=-1, 1], c='blue', s=50)

plt.show()
```

Уже лучше! Распознали 3 кластера, но все равно есть ощущение, что многие шумы были учтены.
Уменьшим eps.

```python
new_predicted = DBSCAN(eps=0.3, min_samples=30).fit_predict(X)
```

```python
plt.figure(figsize=(12,9))
plt.scatter(X[:, 0], X[:, 1], c=new_predicted, s=50)
```

```python
plt.scatter(X[new_predicted==-1, 0], X[new_predicted==-1, 1], c='blue', s=70, marker='x')
plt.scatter(X[new_predicted!=-1, 0], X[new_predicted!=-1, 1], c='green', s=50)
```
